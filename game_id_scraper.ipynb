```python
url_l2m ='https://official.nba.com/2022-23-nba-officiating-last-two-minute-reports/'

def game_id_scraper(url_l2m_season):
    
    # Request data from url
    r = requests.get(url_l2m_season)
    
    # Create 
    data = soup(r.text,'lxml')
    
    # Grab the a tag
    data_list = data.select('.entry-content a')

    # Create an empty list to append links to
    links_l2m = []

    # Pull hrefs from the data_list
    for a in data_list:
        links_l2m.append(a['href'])

    # Create empty game ID list to append to
    game_id_list = []

    # Go through each list in item and slice [51:61] 
    for a in links_l2m:
        game_id  = a[51:61]

        # Test to see if the sliced string could qualify as a numeric and append if yes 
        if game_id.isnumeric():
            game_id_list.append(game_id)

    url_list = []
    for game_id in game_id_list:
        url_list.append(f'https://official.nba.com/l2m/json/{game_id}.json')

    return url_list
```

```python
l2m_reports_21_22 = game_id_scraper('https://official.nba.com/2021-22-nba-officiating-last-two-minute-reports/')
```

```python
l2m_reports_20_21 = game_id_scraper('https://official.nba.com/2020-21-nba-officiating-last-two-minute-reports/')
```
